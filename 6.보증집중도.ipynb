```python
import pandas as pd
cd = pd.read_csv('상권_보증집중도_ver2.csv',encoding='euc-kr')
```

```python
cd
```

```python
#cd.rename(columns={'NUMPOINTS':'사업체수2'},inplace=True)
```

```python
#cd['보증집중도'] = cd['사업체수2']/cd['상가수']
cd['보증집중도'] = cd['보증수']/cd['상가수']
cd['보증집중도'] = cd['보증집중도'].fillna(0)
```

```python
cd[cd['보증집중도']>1].sort_values('보증집중도',ascending=False)


len(cd[cd['보증집중도']>1])
```

```python
cd[cd['보증집중도']>1]
```

```python
#cd_rank = cd[~((cd['상가수']==0) | (cd['상가수']<cd['사업체수']))]
cd_rank=cd
```

```python
bins=[]
for i in reversed(range(1,11)):
    #bins.append(str(i)+'등급')
    bins.append(i)
```

```python
bins
```

```python
pd.cut(cd_rank["보증집중도"], 10, right=False)
```

```python
cd_rank['보증집중등급'] = pd.qcut(cd_rank["보증집중도"], 10, labels=bins)
```

```python
cd_rank['보증집중등급'].value_counts()
```

```python
cd_rank.rename(columns={'다중지?':'다중지역정보','업종정?':'업종정보','상권코?':'상권코드','사업체?':'사업체수'},inplace=True)
```

```python
cd_guarantee = cd_rank[['상권코드','보증집중도','보증집중등급','상가수','보증수']]
```
