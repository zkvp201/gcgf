```python
import pandas as pd
import os
```

```python
df = pd.read_csv('수정/업종 및 분류_수정.csv',encoding='euc-kr')
```

```python
df = df.dropna(subset=[' 인허가 입력 업종 '])
```

```python
df = df[df.columns[:5]]
```

```python
dic = { name:value for name, value in zip(df[' 인허가 입력 업종 '], df['GCGF 업종분류']) }
```

```python
file_list =os.listdir('수정')

csv_list=[]
for a in file_list:
    if a[-4:] ==".csv":
        csv_list.append(a)
```

```python
csv_list = csv_list[:-1]
```

```python
csv_list[0]
```

```python
for a in csv_list:
    df = pd.read_csv('수정/'+a,encoding='cp949')
    for key,value in dic.items():
        if(a[23:-9]==key):
            if(value=='서비스'):
                df.to_csv('수정/서비스/'+a,encoding='euc-kr',index=False)
            elif(value=='도소매'):
                df.to_csv('수정/도소매/'+a,encoding='euc-kr',index=False)
            elif(value=='제조'):
                df.to_csv('수정/제조/'+a,encoding='euc-kr',index=False)
            elif(value=='음식숙박'):
                df.to_csv('수정/음식숙박/'+a,encoding='euc-kr',index=False)
            elif(value=='기타'):
                df.to_csv('수정/기타/'+a,encoding='euc-kr',index=False)
```

```python
#평균업력에필요한 데이터를 분류한 폴더에서 선택하여 미리 집어넣음 (서비스,도소매,음숙업)
sum_list =os.listdir('수정/평균업력')
```

```python
data_set=pd.DataFrame()
for a in sum_list:
    df=pd.read_csv('수정/평균업력/'+a,encoding='euc-kr')
    data_set = pd.concat([data_set,df],axis=0)    
```

```python
data_set.to_csv('인허가데이터_평균업력.csv',encoding='euc-kr',index=False)
```
